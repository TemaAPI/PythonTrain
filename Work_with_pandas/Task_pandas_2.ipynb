### Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:
* оценка 2 и ниже — низкий рейтинг;
* оценка 4 и ниже — средний рейтинг;
* оценка 4.5 и 5 — высокий рейтн
г.


Результат классификации запишите в столбец class.


```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
movie_ratings = pd.merge(movies, ratings, on='movieId')
```

```python
movie_ratings.loc[movie_ratings['rating'] <= 2, 'class'] = 'low'
movie_ratings.loc[(movie_ratings['rating'] > 2) & (movie_ratings['rating'] <= 4), 'class'] = 'medium'
movie_ratings.loc[(movie_ratings['rating'] > 4) & (movie_ratings['rating'] <= 5), 'class'] = 'high'
```

Используйте файл keywords.csv.
Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'бец region.


Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
```

```python
keywords = pd.read_csv('keywords.csv')
```

```python
keywords['region'] = 'undefined'
```

```python
for region, cities in geo_data.items():
        for city in cities:
            keywords.loc[keywords['keyword'].str.contains(city), 'region'] = region
```
