### Задание 1

Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
merged = pd.merge(movies, ratings, on='movieId')
```

```python
filtered = merged[merged['rating'] == 5.0]
```

```python
counts = filtered.groupby('title')['rating'].count().reset_index()
```

```python
sorted_counts = counts.sort_values('rating', ascending=False)
```

```python
most_5_star_ratings = sorted_counts.iloc[0]['title']
```

```python
print("Фильм, которому было выставлено больше всего оценок 5.0:", most_5_star_ratings)
```

### Задание 2

По данным файла power.csv (находится в “Материалы к домашнему заданию”) посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
transactions = pd.read_csv('transactions.csv')
power = pd.read_csv('power.csv')
```

```python
transactions.head(30)
```

```python
power.head(30)
```

```python
power['country'].unique()
```

```python
categories = [4, 12, 21]
countries = ['Latvia', 'Lithuania', 'Estonia']
filtered = power[(power['country'].isin(countries)) & (power['category'].isin(categories))]
```

```python
filtered = filtered[(filtered['year'] >= 2005) & (filtered['year'] <= 2010)]
```

```python
filtered = filtered[filtered['quantity'] >= 0]
```

```python
total_consumption = filtered['quantity'].sum()
```

```python
print("Суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год:", total_consumption)
```
