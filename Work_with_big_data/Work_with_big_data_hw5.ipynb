```python
import pandas as pd
```

```python
data = pd.read_csv('uplift_synthetic_data_100trials.csv')
```

```python
data.head(25)
```

```python
print(data.describe())
```

```python
groups = data.groupby('treatment_group_key')
```

```python
for name, group in groups:
    print(f"Group {name}:")
    print(group.iloc[:, 3:-3].describe())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, 3:-3], data['conversion'], test_size=0.2, random_state=42)
```

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
print(f"Precision: {precision_score(y_test, y_pred)}")
```

```python
print(f"Recall: {recall_score(y_test, y_pred)}")
```

```python
print(f"F1 score: {f1_score(y_test, y_pred)}")
```
